**Câu 3**: Python là ngôn ngữ biên dịch hay thông dịch? Vì sao?

**Trả Lời** :

_ Python là một ngôn ngữ thông dịch (Interpreted Language). Khi một chương trình Python được thực thi, nó không được biên dịch thành mã máy trực tiếp như các ngôn ngữ biên dịch truyền thống như C++.

_ Quá trình thực thi của một chương trình Python bao gồm các bước sau:
*   **Phân tích cú pháp** (Parsing): Trình thông dịch Python phân tích cú pháp của mã nguồn Python để xác định cú pháp và cấu trúc của chương trình.
*   **Biên dịch thành bytecode**: Mã nguồn Python được biên dịch thành bytecode, một dạng ngôn ngữ trung gian dễ dàng để thực thi. Bytecode được lưu trữ trong các tệp .pyc (Python compiled) hoặc .pyo (Python optimized).
*   **Thực thi bytecode**: Trình thông dịch Python (Python interpreter) đọc và thực thi bytecode từng dòng một. Các câu lệnh trong bytecode được dịch và thực thi ngay lập tức.

_ Do đó, Python được coi là ngôn ngữ thông dịch vì mã nguồn không được biên dịch thành mã máy trước khi thực thi. Mỗi lệnh trong Python được dịch và thực thi tại thời điểm thực thi, cho phép chúng ta kiểm tra và thay đổi mã nguồn trong quá trình chạy.

  _ Số thực là số âm, dương kèm theo phần thập phân.

  _ Trong Python giá trị số thực có thể lưu lớn nhất xấp xỉ 1.8x10^308, các giá trị lớn hơn giá trị này được mô tả bởi chuỗi inf (Infinity). Giá trị số thực nhỏ nhất có thể lưu là 5.0x10^-324, các giá trị nhỏ hơn số này được coi là 0.

  _ Ta cũng có thể in số thực với số lượng chữ số sau dấu phẩy xác định.




```python
# Kiểm tra độ lớn , nhỏ của các số thuộc kiểu dữ liệu số thực
a = 3.5
b = 3e5
c = 1.9e308
d = 5.4e-325
print(a , b , c , d , sep = "\n")
# In số thực với 3 chữ số sau dấu phẩy xác định.
e = 1.23456
print(round(e , 3))
print("%3f" % e)
print('{:.2f}'.format(e))
```

  +) Lưu ý : Khi ta sử dụng toán tử so sánh thì kết quả của phép so sánh sẽ trả về True hoặc False



* **Toán tử logic**

  _ Ta có thể sử dụng các toán tử logic and , or , not để kết hợp nhiều biểu thức so sánh

  +) Toán tử and (Toán tử và) : Trả về kết quả là True nếu các mệnh đề thành phần là True , trả về kết quả False nếu có 1 mệnh đề thành phần là False



```python
a , b , c = 100 , 200 , 300
print((a == b) and (b < c))
print((a < b) and (b < c))
```

  +) Toán tử or (Toán tử hoặc) : Trả về kết quả là True nếu một trong các mệnh đề thành phần là True , trả về kết quả False nếu tất cả mệnh đề thành phần là False

```python
a , b , c = 100 , 200 , 300
print((a > b) or (b < c))
print((a > b) or (a > c))
```

  +) Toán tử not (Toán tử phủ định) : Trả về kết quả là True nếu biểu thức đó là False và ngược lại

```python
a , b = 100 , 200
print(not(a == a))
print(not(a > b))
```



* **Toán tử nhận dạng** : Toán tử nhận dạng dùng để so sánh hai đối tượng chứ không phải hai giá trị

  +) is : Trả về True nếu hai đối tượng là giống nhau (cùng địa chỉ bộ nhớ)
  
  +) is not : Trả về True nếu hai đối tượng là khác nhau (khác địa chỉ bộ nhớ)




```python
a = [1 , 2 , 3]
b = a
c = [1 , 2 , 3]
print(a is b)
print(a is c)
print(b is not c)
```



*  **Toán tử thành viên** : Toán tử thành viên được dùng để kiểm tra xem sự tồn tại của một đối tượng trong list , xâu , tuple
  +) in : Trả về True nếu đối tượng kiểm tra tồn tại
  +) not in : Trả về False nếu đối tượng kiểm tra không tồn tại



```python
s = 'a'
x = 'e'
t = 'abcd'
print(s in t)
print(x not in t)
```
